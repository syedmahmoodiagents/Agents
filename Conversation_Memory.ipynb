```python
import os, getpass
```

```python
os.environ["OPENAI_API_KEY"] = getpass.getpass("OpenAI API Key:")
```

```python
from langchain_openai import ChatOpenAI
```

```python
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from langchain.memory import ConversationSummaryMemory
```

```python
llm = ChatOpenAI(model="gpt-4o-mini")
```

```python
memory = ConversationBufferMemory(return_messages=True)
```

```python
conv = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)
```

```python
response1 = conv.invoke({"input": "Explain Newton's first law in not more than one line."})
```

```python
print("AI:", response1)
```

```python
response2 = conv.invoke({"input": "Can you give me a one real-world example in not more than one line?"})
```

```python
print("AI:", response2)
```

```python
response3 = conv.invoke({"input": "How is it different from the second law in not more than one line?"})
```

```python
print("AI:", response3)
```

```python
memory2 = ConversationSummaryMemory(
    llm=llm, memory_key="chat_history", return_messages=True
)  # uses the LLM to summarize history
```

```python
chain = ConversationChain(
    llm=llm,
    memory=memory2,
    verbose=True
)
```

```python
response = chain.predict(input="I live in Hyderabad and I work in pharma domain.")
```

```python
print(response)
```
